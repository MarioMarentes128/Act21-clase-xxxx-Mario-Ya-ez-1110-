```python
class Youtuber:
    # El método __init__ asigna los atributos al crear el objeto
    def __init__(self, nombre, autenticidad, carisma, creatividad):
        self.nombre = nombre  # Atributo extra para identificar cada objeto
        self.autenticidad = autenticidad
        self.carisma = carisma
        self.creatividad = creatividad

    # Función 1: Entretener
    def entretener(self):
        return f"{self.nombre} está entreteniendo a su audiencia con un nivel de carisma de {self.carisma}."

    # Función 2: Hacer videos
    def hacervideos(self):
        return f"{self.nombre} está grabando un nuevo video usando su {self.creatividad}."

    # Función 3: Ganar dinero
    def ganardinero(self):
        return f"¡Éxito! Gracias a su {self.autenticidad}, {self.nombre} ha monetizado su contenido."

# --- Creación de Objetos (Instancias) ---

# Objeto 1
youtuber1 = Youtuber("Alex el Capo", "Alta", "9/10", "Gaming Pro")

# Objeto 2
youtuber2 = Youtuber("Ter", "Única", "10/10", "Arquitectura Pop")

# --- Ejecución y resultados ---

print(f"--- Datos de {youtuber1.nombre} ---")
print(youtuber1.entretener())
print(youtuber1.hacervideos())
print(youtuber1.ganardinero())

print("\n" + "-"*30 + "\n")

print(f"--- Datos de {youtuber2.nombre} ---")
print(youtuber2.entretener())
print(youtuber2.hacervideos())
print(youtuber2.ganardinero())

print("Programa termidado")
print("Crado sofware por: Mario Yañez NL = 53")
```
